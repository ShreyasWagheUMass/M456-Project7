```python
from transformers import AutoTokenizer
```

The next code defines a function `get_tokenizer` which takes a model name ("bert", "roberta", or "gpt2") as input and returns a pretrained tokenizer for that model. 



```python
def get_tokenizer(model_name):
  if model_name == "bert":
    return AutoTokenizer.from_pretrained('bert-base-cased')
  elif model_name == "roberta":
    return AutoTokenizer.from_pretrained('roberta-base')
  elif model_name == 'gpt2':
    return AutoTokenizer.from_pretrained('gpt2')
  else:
    print("ERROR: Unexpected model name '{}'".format(model_name))
```

So, `tokenizer` is a pretrained tokenizer for BERT.

```python
tokenizer = get_tokenizer('bert')
tokenizer2 = get_tokenizer('roberta')
tokenizer3 = get_tokenizer('gpt2')
```

The tokenizer's function `tokenize`, will return the tokenization of the input string. 

```python
tokenizer.tokenize(" The dog fetched the stick.")
```

Notes for Problem I:
- For consistency of comparison of tokenizers, be sure to **include a space before each word you want to tokenize**.
- The leading symbol `Ä ` indicates the start of a new word token for some tokenizers
- The leading sequence `##` indicates the continuation of a word token for some tokenizers 


```python
def token(x):
  print('bert: {}'.format(tokenizer.tokenize(x)))
  print('roberta: {}'.format(tokenizer2.tokenize(x)))
  print('gpt2: {}'.format(tokenizer3.tokenize(x)))
```
